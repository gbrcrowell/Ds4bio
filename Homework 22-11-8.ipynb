```python
import pandas as pd, numpy as np, seaborn as sns
from scipy.stats import norm
pheno = pd.read_table("arabidopsis_birds/arabmagic/arabmagic_pheno.csv", sep=",",skiprows=3,index_col=0)[~pd.isnull(pheno['fruit_length'])].loc[:,'fruit_length']
# imports and pheno creation
```

```python
model1 = norm(loc = 15, scale = 1)
model2 = norm(loc = 15, scale = 2)
model3 = norm(loc = 15, scale = 3)
pheno2 = pheno[~pd.isnull(pheno['fruit_length'])]
pd.read_table("arabidopsis_birds/arabmagic/arabmagic_pheno.csv", sep=",",skiprows=3,index_col=0)[~pd.isnull(pheno['fruit_length'])].loc[:,'fruit_length']
# model creation and creating a pheno called pheno2 that include only those rows that include a value for fruit length
```

```python
df_mod1 = pd.DataFrame(np.log(model1.pdf(pheno2['fruit_length'])) * -1).rename(columns = {0:'model 1'})
# creates a dataframe of the probability of each fruit length based on our model (model1 in this case)
# to do this it finds the probability of each fruit length value with .pdf then I took the log-1 of each.
# we then take that and put in a dataframe rename the 1 column as model 1

df_mod2 = pd.DataFrame(np.log(model2.pdf(pheno2['fruit_length'])) * -1).rename(columns = {0:'model 2'})
# creates a dataframe of the probability of each fruit length based on our model (model2 in this case)
# to do this it finds the probability of each fruit length value with .pdf then I took the log-1 of each.
# we then take that and put in a dataframe rename the 1 column as model 2

df_mod3 = pd.DataFrame(np.log(model3.pdf(pheno2['fruit_length'])) * -1).rename(columns = {0:'model 3'})
# creates a dataframe of the probability of each fruit length based on our model (model3 in this case)
# to do this it finds the probability of each fruit length value with .pdf then I took the log-1 of each.
# we then take that and put in a dataframe rename the 1 column as model 3

df_fruit = pd.DataFrame(pheno2['fruit_length'].values).rename(columns = {0:'fruit length'})
#shifs fruit length into a data frame without magic as its index (now it is just numbers 0:len(pheno2)) and returns its column name
# there is an easier way to do this but I don't remember it (brute force it is).

data = pd.concat((df_mod1, df_mod2, df_mod3, df_fruit), axis = 1).set_index('fruit length')
# combines all of our prevously made dataframes with the new index being fruit length
# axis 1 is use because we want to build adding more columns not more rows
# index was set so I didn't have to spesify too much when building the relplot.

sns.relplot(data=data)
# creates a plot from our combined dataframes with the x-axis as fruit length and y axis being probability of each of our models
```
