**Exercise 6.1**: Using a different marker to predict fruit length (Model 2)
1. Use the marker MASC02863 as a predictor variable and do regression predicting fruit length (save this into a variable called `model2`).

```python
to_model = pd.DataFrame({"fruit_length":fruit_length_geno['fruit_length'],
                         "MASC02863": (fruit_length_geno['MASC02863']=='A').astype(int)})
to_model = to_model.loc[ pd.isnull(to_model).sum(axis=1)==0, :]
to_model = sm.add_constant(to_model)
# We create a data frame called to model, with three columns. The first being fruit length, the second being MASC split into ones and zeros for A and B, and the third being a constant.
# we us isnull to remove any lines that have no value

print('MASC06116',to_model.shape)
X_marker = to_model.loc[:,['const','MASC02863']]
y_marker = to_model['fruit_length']
model2 = sm.OLS(y_marker, X_marker).fit()
# sets an x and y values that will be used in our linear model then plugs them in
# our x values will consist of MASC (which will change between one and zero), and a constant.
# our y values wil consist of fruit length
```

2. Write down the linear model including the coefficients of the model.

```python
print(f'The linear Model is: \nmean_fruit_length = ({model2.params[1]} * MASC02863) + {model2.params[0]}')
```

3. Calculate the predicted mean fruit lengths and the RMSE

```python
predicted_values = model2.predict()
RMSE = np.sqrt((model2.resid ** 2).mean())
# PEMDAS:
# ** 2 squares the value
# .mean takes the mean
# np.sqrt takes square root

print(f'The predicted values of mea fruit length are: \n\t{predicted_values}\nThe RSME is:\n\t{RMSE}')
```

4. Is this marker a good predictor of fruit length? Why? Compare it to `model_marker` using at least 2 metrics.

```python
to_model = pd.DataFrame({"fruit_length":fruit_length_geno['fruit_length'],
                         "MASC06116": (fruit_length_geno['MASC06116']=='A').astype(int)})
to_model = to_model.loc[ pd.isnull(to_model).sum(axis=1)==0, :]

to_model = sm.add_constant(to_model)

print('MASC06116',to_model.shape)
X_marker = to_model.loc[:,['const','MASC06116']]
y_marker = to_model['fruit_length']
model_marker = sm.OLS(y_marker, X_marker).fit()
# making model marker, code taken from Notebook 20

print(f'model marker: {model_marker.rsquared} \nmodel2: {model2.rsquared}')
print(f'model marker: {np.sqrt((model_marker.resid ** 2).mean())} \nmodel2: {RMSE}')
```
